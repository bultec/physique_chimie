# Vecteur variation de vitesse

## Document 1 : Méthode de tracé d'un vecteur vitesse $\overrightarrow{v}$
On a représenté ci-dessous différentes positions d’une balle à intervalles de temps courts et égaux au cours de son mouvement.



Le vecteur vitesse   de la balle à la position $M_4$ est considéré comme égal au vecteur vitesse moyenne entre les positions $M_3$ et $M_5$.
On calcule la valeur du vecteur vitesse moyenne entre les positions $M_3$ et $M_5$:
$$v_4=\frac{M_3M_5}{t_5-t_3} $$

$t_3$ et $t_5$ sont les dates proches de passage en $M_3$ et $M_5$.
En fonction de la valeur $v_4$ de la vitesse, on propose une échelle.

On trace le vecteur vitesse $\overrightarrow{v_4}$ à partir de la position $M_4$, tangent à la trajectoire, dans le sens du mouvement. La longueur du segment fléché est proportionnelle à la valeur de la vitesse.

## Document 4 : un peu de langage python
Si `X` représente un tableau de $n$ nombres, l'accès à la case d’indice $i$ du tableau `X` s’effectue à l’aide de l’instruction `X[i]`.

Par exemple :

```python
a = X[i]
```

__Remarque :__ les indices d’un tableau commencent à la valeur 0.

Par exemple:

    X = [3, 5, 2, 7]
    X[1] sera égal à 5

On peut représenter un point ou un vecteur par un couple de valeurs :

    A = (3, 5)

représentera le point A de coordonnées `x = 3` et `y = 5`.
L’abscisse du point A sera accessible par l’expression `A[0]` et son ordonnée par l’expression `A[1]`.

Pour tracer un vecteur, on pourra utiliser la fonction `trace_vecteur`.

Par exemple:

```python
trace_vecteur(A, vA, 'blue')
```

trace le vecteur $\overrightarrow{V_A}$ au point A avec une couleur bleue.
